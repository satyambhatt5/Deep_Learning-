```python
import math
math.e
```

```python
#Implementing Sigmoid 
# 1/(1+e^-x) 


def sigmoid(x):
    return 1/(1+math.e**(-x))
print(f'Sigmoid: {sigmoid(-0.322)}')


#tanh - (e^x - e^-x)/(e^x + e^-x)
def tanh(x):
    num = ((math.e**x) - (math.e**(-x)))
    den = ((math.e**x) + (math.e**(-x)))
    return num/den
print(f"Tanh: {tanh(0.087)}")


def relu(x):
    return max(0,x)
print(f"Relu: {relu(90)}")
```
